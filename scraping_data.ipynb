```python
from random import randint

def gen_ips():
    network=ipaddress.IPv4Network("192.168.1.0/24")
    hosts=[]
    for host in network.hosts():
        hosts.append(host)
    return hosts
def scraping_data():
    dados=[]
    x=0
    y=0
    for data in nasa+eso:
        if 'galax' in data:
          host=hosts[y]
          network=ipaddress.IPv4Network(host)
          dados.append(requests.get(data).content)
          print(data)
          x+=1
          if x==8:
              y+=1
          sleep(randint(1,3))
    return dados
```

```python
hosts=gen_ips()
```

```python
dados=scraping_data()
```

```python
imgsnasa=[]
for dado in dados:
  info=dado.decode('utf-8')
  if 'nasa' in info:
    while info != '':
      imgsnasa.append(info.partition('"href":"')[2].partition('"')[0])
      info=info.partition('"href":"')[2].partition('"')[2]
print(imgsnasa)
```

```python
nasa_galaxies=[]
for img in imgsnasa:
  if '.json' in img:
    img=requests.get(img).content
    try:
      imgs=json.loads(img)
      for img in imgs:
        nasa_galaxies.append(img)
    except:
      nasa_galaxies.append(img)
  elif img != '':
    nasa_galaxies.append(img)
  else:
    pass
```

```python
for data in nasa_galaxies:
  with open ('sources_nasa.txt','a') as f:
    try:
      f.write(data+'\n')
    except:
      pass
```

```python
# opening the file in read mode
sources_nasa = open("sources_nasa.txt", "r")
# reading the file
nasa = sources_nasa.read()
nasa_list = nasa.split("\n")
  
sources_nasa.close()
```

```python
nasa_galaxies_info=[]
i=0
for img_url in nasa_list:
  try:
    if '~small.jpg' in img_url and 'video' not in img_url:
      starname=img_url.partition('image/')[2].partition('/')[0]
      nasa_galaxies_info.append({"star":starname,"url":img_url,"metadata":""})
    elif '.json' in img_url and 'video' not in img_url:
      metadata=img_url
      nasa_galaxies_info[i]['metadata']=img_url
      i+=1
  except:
    pass
```
