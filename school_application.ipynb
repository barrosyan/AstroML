```python
#TESTING:
jsonfiles=[]
urls=[]
with open (info,'r') as f:
  jsoninfo=f.readlines()
for info in jsoninfo:
  if '.json' in info:
    jsonfiles.append(info)
  elif 'small' in info:
    urls.append(info)
```

```python
name=input('Hi Student! Welcome to the Spatial View Project!\nTell me the code of the galaxy you want to know more...')
for url in urls:
  if name in url:
    print(f'Encontrei a galáxia no nosso banco de dados!\nAguarde um instante para visualizar uma foto reduzida da galáxia.\nPara mais referências, veja no link: {url}')
```

```python
filename=nasa+'/'+name+'.jpg'
image=cv2.imread(filename,0)
plt.imshow(image)
```

```python
for url in jsonfiles:
  if name in url:
    url='https://'+url.partition('http://')[2]
    data=requests.get(url.replace(' ','').replace('\n','')).content
    data=literal_eval(data.decode('utf-8'))
    description=data['AVAIL:Description']
    keywords=data['AVAIL:Keywords']
    location=data['AVAIL:Location']
    id=data['AVAIL:NASAID']
    photographer=data['AVAIL:Photographer']
    secondary_creator=data['AVAIL:SecondaryCreator']
    title=data['AVAIL:Title']
    shape=data['Composite:ImageSize']
    megapixels=data['Composite:Megapixels']
    colorspace=data['EXIF:ColorSpace']
    components=data['EXIF:ComponentsConfiguration']
    creation_data=data['EXIF:CreateDate']
    filesize=data['File:FileSize']
    filetype=data['File:FileType']
    credit=data['XMP:Credit']    
    source=data['XMP:Source']

    print(f"""'\033[1mHere is a brief report abour this galaxy:\033[0m\n\033[1mDescription:\033[0m {description}\n\033[1mKeywords:\033[0m {keywords}\n\033[1mLocation:\033[0m {location}\n\033[1mID:\033[0m {id}\n\033[1mPhotographer:\033[0m {photographer}\n\033[1mSecondary Creator:\033[0m {secondary_creator}\n\033[1mTitle:\033[0m {title}\n\033[1mShape:\033[0m {shape}\n\033[1mPixels:\033[0m {megapixels}\n\033[1mColor:\033[0m {colorspace}\n\033[1mComponents:\033[0m {components}\n\033[1mCreation Data:\033[0m {creation_data}\n\033[1mFile Size:\033[0m {filesize}\n\033[1mFile Type:\033[0m {filetype}\n\033[1mCredit:\033[0m {credit}\n\033[1mSource:\033[0m {source}""")
```
