```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time

chrome_options = Options()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode
chrome_options.add_argument("--disable-gpu")  # Disable the GPU usage
chrome_options.add_argument("--no-sandbox")  # Bypass OS security model

driver = webdriver.Chrome(options=chrome_options)

names=["Irregular+galaxies","Hubble+Irregular+galaxies","Faulkes+Irregular+galaxies",
       "NASA+Irregular+galaxies","ESA+Irregular+galaxies","ISRO+Irregular+galaxies",
       "JAXA+Irregular+galaxies","CNSA+Irregular+galaxies","ROSCOSMOS+Irregular+galaxies"]

urls=[]
for name in names:
    url = f'https://www.google.com/search?q={name}&tbm=isch'
    driver.get(url)
    for x in range(0,10):
        driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
        time.sleep(2)
        soup=BeautifulSoup(driver.page_source,'html.parser')
        url=str(soup).partition('"https:')[2].partition('.jpg')[0]+'jpg'
        while url !='':
            urls.append(url)
            url=str(url).partition('"https:')[2].partition('.jpg')[0]+'jpg'
        url=str(soup).partition('"https:')[2].partition('.png')[0]+'png'
        while url !='':
            urls.append(url)
            url=str(url).partition('"https:')[2].partition('.png')[0]+'png'
urls
```

```python
request=requests.get(urls[0]).content
```

```python
links=[]
soup=BeautifulSoup(request,'html.parser')
a=soup.find_all('a',href=True)
for link in a:
  try:
    request=requests.get(link['href']).content
    soup=BeautifulSoup(request,'html.parser')
    a=soup.find_all('a',href=True)
    for link in a:
      if 'png' in link['href'] or 'jpg' in link['href']:
        links.append(link['href'])
  except:
    pass
```
